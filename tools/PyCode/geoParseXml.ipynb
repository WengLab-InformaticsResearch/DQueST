```python
import xml.etree.cElementTree as ET
import pandas as pd
from os.path import isfile, isdir, join
from os import listdir
from multiprocessing.dummy import Pool as ThreadPool 
```

```python
def initDf ():
    df = pd.DataFrame({'nct_id' : [],'city':[],'state':[],'zip':[],'country':[]})
    return df
```

```python
def xml2row(file):
    df = initDf()
    parsed = {}
    parsed['nct_id'] = None
    for event, elem in ET.iterparse(file):
        if elem.tag == "nct_id":
            parsed['nct_id'] = elem.text
            parsed['city'] = None
            parsed['state'] = None
            parsed['zip'] = None
            parsed['country'] = None     
        if elem.tag == "address":
            parsed['city'] = None
            parsed['state'] = None
            parsed['zip'] = None
            parsed['country'] = None   
            for k in elem.getchildren():
                if k.tag == "city":
                    parsed['city'] = k.text
                if k.tag == "state":
                    parsed['state'] = k.text
                if k.tag == 'zip':
                    parsed['zip'] = k.text
                if k.tag == "country":
                    parsed['country'] = k.text
            df = df.append(parsed,ignore_index = True)
    return df
```

```python
def printSubDf (directory):
    onlyXML = [f for f in listdir(directory) if isfile(join(directory, f)) & f.endswith('xml')]
    df = initDf()
    for f in onlyXML:
        file = join(directory, f)
        row = xml2row(file)
        df = df.append(row,ignore_index = True)
    return df
```

```python
def processAll (directory):
    df = initDf()
    onlyNctDir = [join(directory,d) for d in listdir(directory) if isdir(join(directory,d)) & d.startswith('NCT')]
    #pool = ThreadPool(8) 
    #results = pool.map(printSubDf, onlyNctDirTest)
    for subDir in onlyNctDir:
        print(subDir)
        df1 = printSubDf(subDir)
        df = df.append(df1,ignore_index = True)
    return df
```

```python
directory = '../../resource/AllPublicXML/'
res = processAll(directory)
```
